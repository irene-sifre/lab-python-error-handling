````python
# 1. Define the function for initializing the inventory with error handling
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory
````

````python
# 2. Modify the calculate_total_price function to include error handling
def calculate_total_price(product):
    valid_price = False
    while not valid_price:
        try:
            price = float(input(f"Enter the price per {product}: $"))
            if price < 0:
                raise ValueError("Price cannot be negative. Please enter a non-negative value.")
            valid_price = True
        except ValueError as error:
            print(f"Error: {error}")
    return price
````

````python
 #3. Modify the get_customer_orders function to include error handling
def get_customer_orders(inventory):
    orders = {}
    valid_order_count = False
    
    while not valid_order_count:
        try:
            num_orders = int(input("How many different products would you like to order? "))
            if num_orders < 0:
                raise ValueError("Number of orders cannot be negative. Please enter a valid number.")
            valid_order_count = True
        except ValueError as error:
            print(f"Error: {error}")
    
    for _ in range(num_orders):
        valid_product = False
        while not valid_product:
            product_name = input("Enter the product name: ")
            if product_name in inventory:
                valid_quantity = False
                while not valid_quantity:
                    try:
                        quantity = int(input(f"Enter the quantity of {product_name} you want to order: "))
                        if quantity < 0:
                            raise ValueError("Quantity cannot be negative. Please enter a valid quantity.")
                        elif quantity > inventory[product_name]:
                            print(f"Insufficient stock for {product_name}. Available quantity: {inventory[product_name]}")
                        else:
                            orders[product_name] = quantity
                            valid_quantity = True
                            valid_product = True
                    except ValueError as error:
                        print(f"Error: {error}")
            else:
                print(f"Product '{product_name}' is not available in the inventory. Please enter a valid product name.")
    
    return orders
````

````python
# 4: Test your code by running the program and deliberately entering invalid quantities and product names. Make sure the error handling mechanism works as expected.
def main():
    products = ["apple", "banana", "orange"]
    
    # Initialize the inventory with error handling
    inventory = initialize_inventory(products)
    
    # Get customer orders with error handling
    orders = get_customer_orders(inventory)
    
    # Calculate total price for each product ordered
    total_cost = 0
    for product, quantity in orders.items():
        price = calculate_total_price(product)
        total_cost += price * quantity
    
    print(f"\nTotal cost of your order is: ${total_cost:.2f}")

if __name__ == "__main__":
    main()
````
